# Definitions
- **Data** means information stored in the form of symbols - 1s and 0s, other digits, letters, special characters, etc. It can be collected, measured, analysed and processed further.


- **Server** is a combination of hardware and software responsible for storing, managing, and processing large amounts of data. E.g. web servers, database servers, FTP servers



- **Data connectivity** regards the ability to connect clients and servers, securing the swift and voluminous transfer of information between them


- **APIs** stands for Application Programming Interfaces are the software tool that makes the above picture a reality. An API is a contract allowing software to share data with each other, i.e. it lets devices and software applications communicate in real time.


- **API Application** a collection of endpoints to which developers can attach, and then extract specific information that can be used by those who are working with the apps


- **Data Asset** describes data that is expected to have some value in the future. E.g. Intellectual property (such as patents or songs), a database with lots of records, websites, code. In short data assets are forms of data or software related to the processing of information


- **Application** is a program designed to perform a specific set of operations for the end user, be it a person or another application. E.g. web browsers, video editing programmes, database management software, etc.

- **Programming** refers to the function the app performs and regards the process of converting inputs into outputs.

Considering all the above definitions an **API** processes requests from the clients to the servers, and in response, brings the requested data from the server. It acts like a messenger.

# Taking a closer look at APIs

From a technical perspective, the great thing about APIs is that sometimes everything you need can be accessed with the ease of just two lines of code.

### Limitations
- An API does require maintenance
- Security is a big problem. APIs must act as a filter (against malicious programmers and overloading) while also guaranteeing quality access
- Private information: Some API services come at a monetary price

# Communication between Software Products and Text Files

- Different pieces of software communicate through text files
- Universal text files are .xml, .csv and .json

# Software Integration Explained

- **Integration** has two distinguishable meanings:
    1. A system or architecture, composed of a few different software products, various programming languages, or other pieces of software which can communicate with each other via APIs, or a common API.
    2. (Classical Def.) A situation where multiple software products can be set up to work as one tool
    
- A third type of integration is a mixture of the above two.

### Example
Imagine a piece of software that could store the data you have on a local level, another product that can use it to make calculations, and a third that plots results beautifully. This is the case with SQL, Python, and Tableau respectively. 

- **Driver:** A software driver can 'drive' a certain software product within another one. It is a piece of software too, but its function is to mediate between two products. E.g. Tableau has built in drivers that can read SQL code, so we can work with the dataset in Tableau. Tableau and Python share the same code and functions through a common API. 

- TabPy is a rest API, from the perspective of Tableau, its server allows you to execute Pythom code remotely and on the fly thanks to a built in driver (Tableau is written in C++). Then we can use the obtained results in Tableau.
