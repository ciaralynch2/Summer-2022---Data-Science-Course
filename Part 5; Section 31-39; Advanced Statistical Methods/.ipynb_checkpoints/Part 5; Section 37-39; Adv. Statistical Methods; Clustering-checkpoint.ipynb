# Introduction
- Cluster analysis is a multivariate statistical technique that groups observations on the basis some of their features or variables they are described by.
- Intuitively speaking, observations in a dataset can be divided into different groups and sometimes this is very useful.
- **The goal of clustering is to maximise the similarity of observations within a cluster and maximise the dissimilarity between clusters.**

Applications:
- Market segmentation
- Image segmentation

Clustering is often used as a preliminary step in all kinds of analysis, it is very useful for exploring the data and identifying patterns.

### Difference between Classification and Clustering


- Classification is about predicting an output category, given input data
- Clustering is about grouping data points together based on similarities among them and difference from others

### Math Prerequisites
**Euclidean N-dimensional distance:**
$$\sqrt{(a_1 - b_1)^2 + ... + (a_n - b_n)^2}$$

**Centroid:**
- A centroid is the mean position of a group of points (aka centre of mass)

# K-means Clustering
1. Choose how many clusters you'd like to have
2. Specify the cluster seeds (A seed is basically a starting centroid, it is chosen at random, or by the data scientist based on previous knowledge of the data)
3. Assign each point to a centroid (this is done based on proximity)
4. Adjust the centroids and repeat the last two steps until you can no longer reassign points

# Example: Clustering with Numerical Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

### Load the data

```python
data = pd.read_csv('3.01. Country clusters.csv')
data
# The latitude and longitude correspond to the geographic centres of the 
# countries
```

### Plot the data

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
# The limits are the natural limits of longitude and latitude
```

### Select the features

```python
x = data.iloc[:, 1:3]
# DataFrame.iloc[row indices, column indices] slices the data frame, given
# rows and columns to be kept
x
```

### Clustering

```python
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
kmeans.get_params()
```

### Clustering results

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
# The first 5 observations are in 1 cluster and the last observation is in
# another cluster
```

```python
dwc = data.copy()
dwc['Clusters'] = identified_clusters
dwc
# dwc = data with clusters
```

```python
plt.scatter(dwc['Longitude'], dwc['Latitude'], c=dwc['Clusters'], cmap = 'rainbow')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

# Selecting the number of clusters
Recall clustering is used to minimize the distance between points in a cluster, and to maximise the distance between clusters. This is done automatically using KMeans.

### WCSS
- Stands for within-cluster sum of squares
- Like SSR and SSE, WCSS is a measure developed within the ANOVA framework
- If we minimize WCSS we have reached the perfect clustering solution\*

Condition:
- If we have 6 observations and 6 clusters then the WCSS is 0, as there is only one point in each cluster. This is not a realistic idea for a dataset with many observations
- If we have 1 cluster for any number of observations, WCSS is at its highest
- There must be some point in between!
- Common sense says we want WCSS to be as low as possible while having a small amount of clusters

# WCSS Numerical & Categorical Data Ex.
### Set up

```python
data = pd.read_csv('3.01. Country clusters.csv')
```

```python
data_map = data.copy()
data_map['Language'] = data_map['Language'].map({'English':0, 'French':1, 'German':2})
```

```python
x = data_map.iloc[:, 1:]
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
id_clusters = kmeans.fit_predict(x)
```

```python
dwc = data_map.copy()
dwc['Cluster'] = id_clusters
```

```python
plt.scatter(dwc['Longitude'],dwc['Latitude'],c=dwc['Cluster'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

### New Stuff

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

### The Elbow Method

```python
cluster = range(1,7)
plt.plot(cluster, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
# We have two options here, 2 or 3 clusters
# A 2 cluster solution would be suboptimal as the leap from 2 to 3 
# is very big
```

# Pros and Cons of K-Means
Pros:
- Simple to understand
- Fast to cluster
- Widely available
- Easy to implement
- Always yields a result (this is also a con as it may be deceiving)

Cons:
1. We need to pick K
2. Sensitive to initialisation (i.e where the seeds are)
3. Sensitive to outliers
4. Produces spherical solutions rather than elliptic (due to using the centroid and Euclidean distance)
5. Standardisation

Remedies:
1. The Elbow Method
2. k-means ++, this is a preliminary algorithm that is run to determine the optimal position of the seeds, *init = kmeans++* is one of the parameters of *kmeans.fit(x)*
3. Remove outliers/ remove one point clusters

# To Standardise and to not Standardise?
That is the question...

- The ultimate aim of standardisation is to reduce the weight of higher numbers.
- It is good practice to standardise before clustering, especially for beginners.

### When Not to Standardise
- If we know one variable is inherently more important than another, standardisation should not be used

# Relationship between Regression and Clustering
- Clustering can help to identify Omitted Variable Bias

### Example


You have a graph detailing the price and size of apartments and there are four clusters; small reasonably priced (A), small expensive (B), big cheap (C), big reasoanbly priced (D).

Usually we would observe that small apartments would be cheaper and big apartments would be more expensive. Maybe B are apartments in the city centre, and C are apartments in the suburbs. Ignoring these we are left with a linear regression.

Clustering here can help identify OVB, e.g. in this case we need to also consider location as a factor on the price, not just size.

# Market segmentation Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

### Load the data

```python
data = pd.read_csv('3.12. Example.csv')
data.head()
```

```python
data.describe()
# Satisfaction is self-reported on a scale from 1-10, it is dicrete

# Loyalty is difficult to measure, there is no widely accepted way
# there are proxies like churn rate, retention rate, or customer 
# lifetime value (CLV)
# It is continuous data with a range (-2.5, 2.5)
```

 ### Plot the data

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction', fontsize = 20)
plt.ylabel('Loyalty', fontsize = 20)
plt.show()
```

### Select the features

```python
x = data.copy()
```

### Clustering

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

### Clustering results

```python
clusters = x.copy()
clusters['clusters_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c=clusters['clusters_pred'], cmap='rainbow')
plt.xlabel('Satisfaction', fontsize = 20)
plt.ylabel('Loyalty', fontsize = 20)
plt.show()
```

Most probably the algorithm ONLY considered satisfaction as a feature, since its values are much larger than loyalty. So we need to standardise.

Whenever we cluster on the basis of a single feature, the result is like the graph above. It's as though there is a vertical line splitting the clusters.

### Standardising

```python
from sklearn import preprocessing as pp
x_scaled = pp.scale(x)
# x_scaled contains the standardised 'Satisfaction' and the same values
# for 'Loyalty' as it was already standardised
```

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
# Given this graph, I think we should use 4 clusters
```

### Explore clustering solutions and select the number of clusters

```python
kmeans_new = KMeans(2)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
clusters_new.head()
```

The dataframe here contains the original values, but the predicted clusters are based on standardized data. We will plot the data without standardising the axes but the solution will be the standardized one.

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c = clusters_new['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

We often choose to plot using the original values for clearer interpretability. Note: the discrepancy we observe here depends on the range of the axes too.

We can also see that both satisfaction and loyalty were taken into account as we no longr have that vertical line.

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
clusters_new.head()
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c = clusters_new['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

It is important to name your clusters
1. **Alienated**, this is the purple cluster, which has low loyalty and low satisfaction
2. **Supporters**, this is the red cluster, which has high loyalty and low satisfaction
3. **Fans**, this is the blue cluster, we want all the points in the graph to be in this cluster
4. **Roamers**, this is the yellow cluster, which has high satisfaction but low loyalty

# How is Clustering Actually Useful
## Types of Analysis
- Exploratory
- Confirmatory
- Explanatory

### Exploratory
- Get acquainted with the data
- Search for patterns
- Plan for what is next

Data visualisation, descriptive statistics, and clustering are great ways to get acquainted with the data without explicitly trying to explain anything.

No strict definition of the other two, their aim is to:
- Explain a phenomenon
- Confirm a hypothesis
- Validate previous research
- This is where we would usually use hypothesis testing and regression analysis

### Clustering
- Clustering can be used for all types of analysis, but it is most commonly used for exploratory analysis. E.g. the market segmentation example was exploratory
- Clustering can also be used to confirm past beliefs, this is less used as classification is more used for this

### Recall Market Segmentation Example
- We want to increase the customer satisfaction of Supporters and increase the loyalty of Roamers.
- By increasing customer satisfaction we may also have some people from the Alienated cluster moving to the Roamers cluster
- In the same way an increase in customer loyalty may have some people from the Alienated cluster to move to the Supporters cluster.


- It doesn't make sense to address the Alienated cluster first as they are the furthest away from the Fans cluster, by default they are not even likely to enter the shop as they are dissatisfied and disloyal.


- Another strategy is to gather as much data on our Fans cluster as possible, and then find people of similar age, interests, income, etc. who have not heard of the business and advertise to them.

# Types of Clustering
- Flat (e.g. K-means)
- Hierarchical

### Hierarchical
- Example: Taxonomy of the Animal Kingdom
- Animal
    - Fish
    - Mammals
    - Birds
        - Flying
        - Flightless
            - Penguins
            - Emu
            - Kiwi

This tree structure indicates hierarchical clustering.

#### Types:
- Agglomerative (Bottom up)
    - The above example is agglomerative
- Divisive (Top down)

# Dendrogram


The distance between the links shows the similarity between observations

### Pros
1. Hierarchical clustering shows all the possible linkages between clusters
2. We understand the data much, much better
3. No need to preset the number of clusters (like with k-means)
4. Many methods to perform hierarchical clustering (Ward)

### Cons
1. Scalability, not useful for large numbers of observations
