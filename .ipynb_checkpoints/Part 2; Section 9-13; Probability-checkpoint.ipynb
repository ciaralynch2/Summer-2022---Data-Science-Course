$\color{red}{\LARGE{\text{General Definitions}}}$
- **Probability:** The likelihood of an event occurring
- **Independent:** The likelihood of one event occurring does not affect the likelihood of a separate event occurring P(A $\cap$ B) = P(A) x P(B)
- **Expected Values:** E(x) = $\sum^{n}_{i}P(x_i)x_i$
- **Complements:** The complement of an event is everything an event is not, $A + A^c$ = Sample Space, $A^c = A'$. $P(A') = 1 - P(A)$

*To construct an interval with the highest probability it should be constructed around the* **expected value**

$\color{red}{\LARGE{\text{Combinatorics}}}$

$\color{blue}{\Large{\text{Permutations}}}$
- The number of different possible ways we can arrange **all** elements within the sample space.
- No repetitions
- The number of permutations is $P_n = n!$

### Factorials
Negative numbers don't have a factorial and $0! = 1$
$$(n+k)! = n!(n+1)...(n+k)$$
$$(n-k)!= \frac{n!}{(n-k+1)(n-k+2)...(n-k+k)}$$
$$\frac{n!}{k!} = (k+1)(k+2)...(n)$$

$\color{blue}{\Large{\text{Variations}}}$
- Variations does not take into account double counting elements

### With Repetitions
This is the total number of ways we can pick and arrange **some elements of a given set (not necessarily all)** with repetition.
$$\bar{V}(n,p) = n^p$$
- $\bar{V}(n,p)$ = the number of variations with repetition when picking p-many elements out of n-elements  
- $n$ = the total number of elements we have available 
- $p$ = the number of positions we need to fill
The intuition behind this is that we have n-options for the first position and we still have n-options for the second as repetition is allowed, this idea continues, meaning every p-position has n-options.

### Without Repetitions
This is the total number of ways we can pick and arrange **some elements of a given set (not necessarily all)** without repetition
$$V(n,p) = \frac{n!}{(n-p)!}$$
- $V(n,p)$ = the number of variations without repetition when picking p-many elements out of n-elements 
- $n$ = the total number of elements we have available 
- $p$ = the number of positions we need to fill
The intuition behind this is that we have n-options for the first element, then (n-1) options for the second since we cannot repeat values. This continues to give $(n)(n-1)...(n-p+1) = \frac{n!}{(n-p)!}$

$\color{blue}{\Large{\text{Combinations}}}$
- The number of different ways we can pick certain elements of a set
- All the different permutations of a single combination are different variations
- In Combinations the order of the elements is irrelevant!
Example:
If we choose Sarah, Alex, and Dave (SAD) this is the same as choosing SDA, ASD, ADS, DAS, DSA, i.e. any of the 6 permutations of this combination is a different variation, but **NOT** a different combination. (Permutation is arrange, Combination is pick, Variation is pick **and** arrange)

### Without Repetitions (see Extra for with)
The number of combinations without repetition equals the number of variations (without repetition) over the number of permutations
$$C(n,p) = \frac{V(n,p)}{P(p)} = \frac{n!}{p!(n-p)!}$$

### Symmetry of Combinations
- $^nC_p = {n\choose p} = \frac{n!}{(n-p)!}$
- Unlike permutations, choosing more elements can result in fewer combinations. (Recall Pascal's Triangle e.g. ${10 \choose 3} = {10 \choose 7}$)
$$^nC_p = ^nC_{n-p}$$
- Example: Picking 4 fruits out of 6 is the same as picking which 2 fruits of the 6 to leave behind
- This symmetry is used to avoid unnecessarily calculating large factorials

$\color{red}{\LARGE{\text{Extra}}}$

$\color{blue}{\Large{\text{Combinations with Repetition}}}$

In special cases we can have repetition in combinations, this requires a different formula.
$$\bar{C}(n,p) = \frac{(n+p-1)!}{(n-1)!p!}$$
- $\bar{C}(n,p)$ = combinations with repetition
- n = total number of elements in the sample space
- p = number of elements we need to select

### Application
Combinations with repetition is used when the events we are dealing with have sufficient quantity of each of the distinct values in the sample space.

**Example:** Toppings on pizza. We can order *extra* of any given topping so repetition is allowed. But the order of the toppings is still irrelevant, so this is not variations.

Imagine you order a 3-topping pizza and they only offer 6 toppings; chicken(c), onions(o), green peppers(g), mushrooms(m), pepperoni(p), bacon(b). The pizza can have 3 different toppings, or you can repeat a topping up to 3 times. The different types of pizza you can have are; 3 of the same topping, 3 different toppings, 2 different toppings with double the amount of 1.

The methodology used for such combinations is a bit abstract, each type of pizza is represented with a special sequence of 0s and 1s. To do this we need to select a specific order for the available ingredients.

To construct the sequence for each unique type of pizza we follow two rules as we go through the ingredients in the order stated above.

1. If we want none/no more of a certain topping we write a 0 and move to the next topping
2. If we want to include a topping we write a 1 **and stay on the same topping**
- If we want extra we write a 1 and **stay on the same topping**, we only move on to the next topping after a 0



### Relationship between combinations without and with repetition

#### ALL SEQUENCES MUST END IN 0
This is the condition for the sequence to end, so it is only the first 8 elements in the sequence that can take different values. Each pizza is characterised by the positions of the three 1s in the sequence. 

**Since only 8 positions in the sequence can take a value of 1, then the number of different pizzas would be the combination of any 3 of the 8 positions**

This idea means we can transform combinations *with* repetition to combinations *without* repetition
$$\bar{C}(6,3) = C(8,3)$$
We can use this to create a general formula. 
- *3* represents the amount of toppings we need to pick, so it is still equivalent to p.
- *8* represents the number of positions we ahve available for the 1s. We had 9 positions (3 1s and 6 0s), but we knew the last position had to be 0, and so could not contain a 1. So we had *n+p-1* positions that could contain a 1
$$\bar{C}(n,p) = C(n+p-1,p) = \frac{(n+p-1)!}{((n+p-1)-p)!p!} = \frac{(n+p-1)!}{(n-1)!p!}$$
