# Simple Linear Regression - Minimal Example
### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()
```

### Generate random import data to train on

```python
obs = 10000
xs = np.random.uniform(-10, 10, (obs, 1))
zs = np.random.uniform(-10, 10, (obs, 1))

print(xs.shape)
print(zs.shape)
inputs = np.column_stack((xs, zs))
print(inputs.shape)

# np.random.uniform(low, high, size) draws a random value from the interval
# (low, high) where each number has an equal chance to be selected
```

### Create the targets we will aim at

In this example we will use fake targets to be sure there is a linear relationship. The algorithm will have to figure out the preset relationship on its own

We will take a completely arbitrary function, e.g f(x,z) = 2x - 3z + 5 + noise
Conceptually the algorithm must LEARN this is the function.

Noise is introduced to randomise the data a bit. Real data always contains noise, it is never perfect.

```python
noise = np.random.uniform(-1, 1, (obs, 1))
targets = 13*xs +7*zs -12 + noise
print(targets.shape)
```

### Plot the training data

```python
# In order to use the 3D plot, the objects should have a certain shape, so
# we reshape the targets. The proper method to use is reshape 
targets = targets.reshape(obs,)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')

# You can fiddle with the azim parameter to plot the data from different 
# angles. 
ax.view_init(azim=100)
plt.show()

targets = targets.reshape(obs, 1)
```

### Initialize variables

```python
init_range = 0.1
# Our initial weights and biases will be picked randomly from the interval
# [-0.1, 0.1]

weights = np.random.uniform(-init_range, init_range, size=(2,1))
biases = np.random.uniform(-init_range, init_range, size = 1)
```

### Set a learning rate

```python
rate = 0.02
```

### Train the model

```python
for i in range(100):
    outputs = np.dot(inputs, weights) + biases
    deltas = outputs - targets
    loss = np.sum(deltas ** 2) / (2*obs)
    print(loss)
    deltas_scaled = deltas / obs 
    weights = weights - rate*np.dot(inputs.T, deltas_scaled)
    biases = biases - rate*np.sum(deltas_scaled)
    
    
# The loss variable gives us the mean loss, we also divide by 2 by 
# convention. This little improvement makes the learning rate independent
# of the number of observations.

# Often when coding matrices, you find the correct way to code it through
# dimensionality checks and compatibility errors. However, transposing
# matrices does not change the info they hold so we can do it freely
```

### Print weights and biases to see if correct

```python
print(weights, biases)
```

### Plot last outputs vs targets
Since they are the last ones at the end of the training, they represent the final model accuracy. The closer this plot is to a 45 degree line, the closer the target and output values are.

```python
plt.plot(outputs, targets)
plt.xlabel('Outputs')
plt.ylabel('Targets')
plt.show()
```
